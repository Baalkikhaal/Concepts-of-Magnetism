# Quasi-classical paramagnetism - $S_{z}$ is quantum

Reference: Amikam Aharoni -- Theory of Ferromagnetism

The treatment of paramagnetism in the quasi-classical framework proceeds on similar lines as that of diamagnetism, with some additional quantum-mechanical **assumptions**, physical **approximations** and statistical **frameworks** namely

- In addition to the spin $\vec{S}$ being quantum mechanical, the component of the spin along the direction of applied field $\vec{S_{z}}$ is also quantum mechanical, in that it can only take values integral or half-integral between $S$ and $-S$, i.e. $2S + 1$ values.

- The approximation that the individual spin interacts with **only** the external applied field $\vec{H}$ and ignoring the spin-spin interaction.

- The energy of interaction is $- \vec{m}\cdot\vec{H}$; with the H along the z-axis the energy values being $-m_{z}H$ and  the alowed values being $-mS, \ldots, +mS $. And that the average behaviour of large number of spins being treated in the classical statistical framework of an ensemble of spins with the average of a statistical quantity like the magnetic moment $\langle m_{z} \rangle$ being

$$
\langle m_{z} \rangle = \frac{\sum m_{z} \exp (m_z \beta H)}{\sum \exp (\beta m_z H)},
$$

where $\beta = \frac{1}{k_B T}$

Using 
\begin{eqnarray}
m_{z} & = & g \mu_{B} S_{z}, \\
\text{where}\qquad \mu_{B} & = & \frac{|e| \hbar}{2 m_{e} c}, \\
\langle m_{z} \rangle & = & \frac{\sum_{n = -S}^{+S} g \mu_{B} n \exp (g \mu_{B} \beta H n)}{\sum_{n = -S}^{+S} \exp ( g \mu_{B} \beta H n)}.
\end{eqnarray}

Further using

\begin{eqnarray}
\xi & = & \exp (g \mu_{B} \beta H), \\
\langle m_{z} \rangle & = & \frac{\sum g \mu_{B} n \xi^{n}}{\sum \xi^{n}}.
\end{eqnarray}

In the scale of spin quantum number $S$, the average component of $S_{z}$ is given by

$$
\langle S_{z} \rangle = \frac {\langle m_{z} \rangle}{g \mu_{b}}  = \frac{\sum n \xi^{n}}{\sum \xi^{n}}.
$$

---

### Algebraic jugglary

The denominator is a geometric series and the numerator is related to the derivative of the denominator with respect to $n$ i.e.

\begin{eqnarray}
\langle S_{z} \rangle & = & \frac{\xi \frac{\partial}{\partial n} (\sum \xi^{n})}{\sum \xi^{n}}, \\
& = & \frac{S(\xi^{S+1} - \xi^{-\overline{S+1}}) - \overline{S+1} (\xi^{S} - \xi^{-S}) }{(1 - \frac{1}{\xi})(\xi^{-S} - \xi^{S + 1})}.
\end{eqnarray}

Using

\begin{eqnarray}
\xi & = & \exp (\eta), \\
\text{where} \qquad \eta & = & \frac{g \mu_{B} H}{k_{B} T}, \\
\langle S_{z} \rangle & = &  \frac {S \{ \sinh (\overline{S+1} \eta) - \sinh (S \eta) \} - \sinh (S \eta)}{\cosh (\overline{S+1} \eta) - \cosh (S \eta)}.
\end{eqnarray}

Using summation properties of hyperbolic functions

\begin{eqnarray}
\sinh A - \sinh B & = & \sinh \Big(\frac{A-B}{2} \Big)\cosh \Big(\frac{A+B}{2} \Big), \\
\cosh A - \cosh B & = & \sinh \Big(\frac{A-B}{2} \Big)\sinh \Big(\frac{A+B}{2} \Big),
\end{eqnarray}

the average component $\langle S_{z} \rangle$ in units of $S$ reads

$$
\frac{\langle S_{z} \rangle}{S} = \frac{2S + 1}{2S} \coth \Big( \frac{\overline{2S+1}}{2} \eta \Big) - \frac{1}{2S} \coth \Big( \frac{\eta}{2}\Big).
$$

Expressing the dependence in terms of the dimensionless variable $\eta S$

\begin{eqnarray}
\frac{\langle S_{z} \rangle}{S} & = & \frac{2S + 1}{2S} \coth \Big( \frac{\overline{2S+1}}{2S} \eta S \Big) - \frac{1}{2S} \coth \Big( \frac{1}{2S} \eta S\Big), \\
                                & = & B_{S}(\eta S), \\
                                & = & B_{S}\Big( \frac{g \mu_B }{k_{B}T} S H \Big), \\
                                \text{where}\\
                       B_{S}(x) & = & \frac{2S + 1}{2S} \coth \Big( \frac{\overline{2S+1}}{2S} x \Big) - \frac{1}{2S} \coth \Big( \frac{1}{2S} x \Big)
\end{eqnarray}

$B_{S}(x)$ is the **Brillouin function**.

## $S = \frac{1}{2}$  special case
For the frequently encountered special case of $S = \frac{1}{2}$, we can use the Brillouin function $B_{\frac{1}{2}} (x)$ or on the other hand rederive the statistcal average expression as the allowed values are only $S_{z} = \pm \frac{1}{2}$.

\begin{eqnarray}
\text{So} \qquad \langle S_{z} \rangle  & = & \frac{\sum_{-\frac{1}{2}}^{\frac{+1}{2}} n \xi^{n}}{\sum_{-\frac{1}{2}}^{\frac{+1}{2}} \xi^{n}}, \\
       & = & \frac{-\frac{1}{2}\exp \Big( - \frac{\eta}{2}\Big) + \frac{1}{2}\exp \Big( \frac{\eta}{2}\Big)}{\exp \Big( - \frac{\eta}{2}\Big) + \exp \Big( \frac{\eta}{2}\Big) },
       & = & \frac{1}{2}\tanh \Big( \frac{\eta}{2}\Big).
\end{eqnarray}

In terms of $S$,

$$\label{langevin}
\langle S_{z} \rangle = S \tanh \Big( \frac{g \mu_{B}}{k_{B}T} S H\Big).
$$

This expression is called the Langevin function. Seemingly, the expressions Brillouin function for $S=\frac{1}{2}$  and the Langevin function appear different, but it can be proved that they are equal.

$$
B_{\frac{1}{2}} (x) = \tanh (x).
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def langevin(x):
    return np.tanh(x)

x = np.arange(-10.0,10.0,0.01)
plt.plot(x,langevin(x),'k',label=r"langevin $= B_{\frac{1}{2}}(x)$")
plt.xlabel(r'$x = \frac{g \mu_{B}}{k_{B} T} SH$')
plt.ylabel(r'$\frac{\langle S_{z} \rangle}{S}$')
plt.legend(loc='best')
plt.grid()
```

```python
def brillouin(x,S):
    factor1 = (2.0*S+1.0)/(2.0*S)
    factor2 = 1.0/(2.0*S)
    return factor1/np.tanh(factor1*x) - factor2/np.tanh(factor2*x)

plt.subplot(211)
plt.plot(x,langevin(x),'k',label='langevin')
plt.xlabel('x')
plt.legend(loc='best')
plt.subplot(212)
plt.plot(x,brillouin(x, 0.5),'b',label=r'brillouin $= B_{\frac{1}{2}}(x)$')
plt.xlabel('x')
plt.legend(loc='best')
```

```python
plt.plot(x, brillouin(x, 0.5),'k',label=r'$B_{\frac{1}{2}}(x)$')
plt.plot(x, brillouin(x, 1.0),'b',label=r'$B_{1}(x)$')
plt.plot(x, brillouin(x, 1.5),'g',label=r'$B_{\frac{3}{2}}(x)$')
plt.plot(x, brillouin(x, 1000),'r',label=r'$B_{\infty}(x)$')
plt.xlabel('x')
plt.legend(loc='best')
plt.grid()
```

## Limiting cases of $x \longrightarrow 0 $ and $x \longrightarrow \infty$

### $x \longrightarrow 0$ : Curie law

We can observe that the brillouin function is a nonlinear function. However the behaviour appears linear close to $x=0$. This corresponds to the experiments at low external applied field $H$.

To extract a low field linear response of the magnetization, let us use the Taylor expansion assuming $x \longrightarrow 0, when$

$$
\cot x = \frac{1}{x} + \frac{x}{3} + \mathcal{O}(x^{3})
$$

So

$$
B_{S}(x) = \frac{S+1}{3S} x + \mathcal{O}(x^{3})
$$

Since $x= \frac{g \mu_{B}}{k_{B}T} SH $, for low applied field $H$

$$
\langle S_{z} \rangle  =  \frac{g \mu_{B} S (S+1)}{k_{B}T} H.
$$

- The **linear response** is valid for most paramagnets at ordinary temperature even for highest possible fields $(\mathcal{O}(10^{2} \text{Oe}))$; it captures the essence of what we mean by a paragmagnet.

- However at higher fields, the deviationsm the linearity is considered by defining  a field dependent susceptibility $\chi (H)$. that equates to the *initial susceptibility* in the linear regime.

\begin{eqnarray}
\chi_{initial} & = & \lim_{H \rightarrow 0}  \frac{\partial}{\partial H} \langle M_{z} \rangle, \\
               & = & \lim_{H \rightarrow 0}  \frac{\partial}{\partial H} g \mu_{B} \langle S_{z} \rangle, \\
               & = & \frac{NS(S+1)(g\mu_B)^{2}}{3k_{B}}\cdot \frac{1}{T}.
\end{eqnarray}

where $N$ is the number of spins per unit volume.

$$
\therefore \qquad \chi_{initial} = \frac{C}{T}.
$$

This agrees with the experimentally observed `Curie law` - i.e. the inverse temperature dependence of susceptibility.

> A key **similarity** between a paramagnet and a ferromagnet is that the susceptibility may depend on field in both cases; but the **difference** is that in paramagnet, $\chi$ **does not** depends on the *history* of H wherease in ferromagnet, it **does**.

### $x \longrightarrow \infty$ by $H \longrightarrow \infty$ : Paramagnetic saturation

Another limiting behaviour of paramagnet constitutes very large applied fields.

$H \longrightarrow \infty$

in which case argurment of $B_{S}(x)$

$ x \longrightarrow \infty$

in which case

$\lim_{x \rightarrow \infty} \coth x = \infty$

$\therefore B_{S}(\pm \infty)  = \pm \infty$,

which signifies that all the spins align parallel to the direction of applied field. This is called *paramagnetic saturation*

- So the effect of field is to only change the direction of individual, fixed magnetic moments.
- It does not change the magnitudes, except for the small *diamagnetic* contribution, which **always** exist.
- Some paramagnets do not exhibit saturation even at largest possible fields $ \sim 10^{6} \text{Oe}$. However, there is another way to achieve such saturation. Observe that the argument in $B_{S}(x)$ contains $\frac{SH}{T}$. So one way to saturate is to perform the experiment at low T.

### $x \longrightarrow \infty$ by $S \longrightarrow \infty$ : Superparamagnetism

- There is onther way to achieve large $x$, when $S$ is large. In normal materials/paramagnets, $S$ is the spin of individual atom/ion and is of the order $1$. We also can reasonably `assume` that the interaction between individual spins is negligible.
- However under certain conditions, these spins couple due to mutual interction and behave as a single $S$ of large magnitude of the order $10^{3} \ \text{to} \ 10^{4}$. This phenomenon is called *superparamagnetism*.
- Superparamagnets saturate, as evident from the Brillouin function, even for small H.


```python
H = np.arange(-5.0,5.0,0.01)
S1 = 0.5
S2 = 10.0
S3 = 100.0
S4 = 1000.0
#let g \mu_{B}/k_{B}T = 1.0

def argument(H, S):
    return H*S



plt.plot(H, brillouin(argument(H, S1), S1),'g',label='S = 1/2')
plt.plot(H, brillouin(argument(H, S2), S2),'b',label='S = 10')
plt.plot(H, brillouin(argument(H, S3), S3),'r',label='S = 100')
plt.plot(H, brillouin(argument(H, S4), S4),'k',label='S = 1000')
plt.title('superparamagnetic saturation')
plt.xlabel(r'$\frac{k_{B}T}{g \mu_{B}} \cdot H$')
plt.legend()
```
